```python
import urllib
import pandas as pd
from tqdm import tqdm_notebook
from keras.applications import vgg19
```

```python
# Read the catalog
df = pd.read_csv('catalog.txt', sep='\t', lineterminator='\r')

# Remove extraneous data we don't need
obsolete_columns = list(df.columns)
obsolete_columns.remove('URL')
df = df.drop(obsolete_columns, axis=1)

# Take out the last data point, which is broken for some reason
df = df.drop([df.shape[0]-1])

# Parse the URLs so that we go straight to the image
for i, url in tqdm_notebook(enumerate(df['URL']), total=df.shape[0]):
    try:
        image_url = url.replace('html', 'art', 1).replace('html', 'jpg', 1)
        df['URL'][i] = image_url
    except AttributeError:
        continue

# For display
df
```

```python
for url in df['URL']:
    urllib.urlretrieve(image_url, '../art_images/%05d.jpg' % i)    
```
